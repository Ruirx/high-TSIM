# An example for High-dimensional sparse settings.

### Note: the optimaization functions are needed reload to avoid confilict with low cases.

```python
import numpy as np
from hTSIM import coef_opt_TSIM
from hTSIM import plot_fun
from hTSIM import gen_pseudo_sample
import matplotlib.pyplot as plt

X, y, fres, Btrue = gen_pseudo_sample(nsample=200, core_size=2, psize=2, nzero=2, sigma=0.1, seed=101)
```

```python
plt.rcParams["figure.figsize"] = [8.0, 6.0]
plt.rcParams["figure.autolayout"] = False
plt.scatter(sorted(fres), y[fres.argsort(0)], color='gray', label = 'observation')
plt.plot(sorted(fres), np.sin(sorted(fres)), color='red', label = 'underlying')
plt.title('generated sample')
plt.legend()
plt.show()
```

```python
core_rank = (2*np.ones(3, dtype=int)).tolist()
Btsim, Betatsim = coef_opt_TSIM(X, y, core_rank, maxitr=100)
```

```python
ytsim = plot_fun(X, y, Btsim, Betatsim, fres, save=True)
```
